```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
attendance = pd.read_csv('../data/2020_game_attendance - Sheet1.csv')
attendance.insert(1, "Name", "")
attendance
```

```python
games = pd.read_csv('../data/games.csv')
games_2020 = games[games['season'] == 2020]
games_2020.head()
```

```python
teams = pd.read_csv('../data/nfl_teams.csv')
teams.head()
```

```python
abbrev_dict = {row['Name']: row['Abbreviation'] for index, row in teams.iterrows()}
abbrev_dict["Washington Football Team"] = "WAS"
abbrev_dict
```

```python
attendance['Name'] = attendance['Tm'].map(abbrev_dict)
```

```python
attendance.info()
```

```python
for column in attendance.columns[5:8]:
    for index, row in attendance.iterrows():
        if pd.isna(row[column]):
            col_name = f"{column}_nan"
            attendance.at[index, col_name] = 1
        else:
            col_name = f"{column}_nan"
            attendance.at[index, col_name] = 0

for column in attendance.columns[8:22]:
    for index, row in attendance.iterrows():
        if pd.isna(row[column]):
            col_name = f"{column}_nan"
            attendance.at[index, col_name] = 1
        else:
            col_name = f"{column}_nan"
            attendance.at[index, col_name] = 0

attendance.head()
```

```python
nan_cols = list(attendance.columns[30:])
for column in nan_cols:
    attendance[column] = attendance[column].astype(int)
attendance.iloc[:, 5:21] = attendance.iloc[:, 5:21].fillna(0)
attendance.head()
```
